```python
import json
from TextRNN import CharRNN
```

```python
fp_poems = '/home/jovyan/Boo/Data/Dem_Debate_Data/poems.json'
```

```python
poems_data = []
with open(fp_poems, encoding='utf-8') as f:
    for line in f:
        poems_data.append(json.loads(line))

text = ''
for idx, line in enumerate(poems_data):
    s = line['text']
    s = s.encode("ascii", errors="ignore").decode()
    text += '‚ñ∂Ô∏è' + s + 'üõë'
```

```python
cRNN = CharRNN(text)
cRNN.build_compile()
```

```python
cRNN.fit(100)
```
